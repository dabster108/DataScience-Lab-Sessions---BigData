```python
from pyspark.sql.functions import col
import pandas as pd
import warnings

warnings.simplefilter(action='ignore', category=UserWarning)

# Rename columns
weather_renamed_columns = weather_table \
    .withColumnRenamed("visib", "visibility") \
    .withColumnRenamed("precip", "precipitation")

# Cast datetime columns to timestamp
for field in weather_renamed_columns.schema.fields:
    if str(field.dataType) in ['TimestampType', 'DateType']:
        weather_renamed_columns = weather_renamed_columns.withColumn(
            field.name, col(field.name).cast("timestamp")
        )

weather_renamed_columns_pd = weather_renamed_columns.toPandas()

# Convert datetime and numeric columns in Pandas
datetime_cols = ['date']  # example, change if needed
for c in datetime_cols:
    if c in weather_renamed_columns_pd.columns:
        weather_renamed_columns_pd[c] = pd.to_datetime(weather_renamed_columns_pd[c], errors='coerce')

numeric_cols = ['temp', 'wind_speed', 'visibility', 'precipitation']  # example
for c in numeric_cols:
    if c in weather_renamed_columns_pd.columns:
        weather_renamed_columns_pd[c] = pd.to_numeric(weather_renamed_columns_pd[c], errors='coerce')

print(weather_renamed_columns_pd.head(5))
```
