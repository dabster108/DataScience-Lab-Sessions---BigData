```python
from pyspark.sql.functions import*
from pyspark.sql.functions import*
```

```python
flights.createOrReplaceTempView("flights_table")
query1 = spark.sql("SELECT year, month, day FROM flights_table")
query1.show()
```

```python
query2 = spark.sql("SELECT year, month, day FROM flights_table WHERE month = 1")
query2.show()

# Convert to Pandas
query2_pd = query2.toPandas()
query2_pd.head()
```
